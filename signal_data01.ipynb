```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
fs = 1000  # Sampling frequency
t = np.linspace(0, 1, fs, endpoint=False)  # Time vector for 1 second

# Generate a clean sine wave (normal signal)
freq = 5  # Frequency of the signal
normal_signal = np.sin(2 * np.pi * freq * t)

# Generate a jammed signal (introducing noise and spikes)
jammed_signal = normal_signal + 0.5 * np.random.normal(size=fs)  # Adding noise
jammed_signal[200:300] += 3  # Introducing a spike

# Plot both signals
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.title("Normal Signal")
plt.plot(t, normal_signal)

plt.subplot(2, 1, 2)
plt.title("Jammed Signal")
plt.plot(t, jammed_signal)
plt.tight_layout()
plt.show()
```
